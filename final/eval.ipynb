```python
import os
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
name = 'test'
```

```python
hist_ss = pd.read_csv(os.path.join('metrics', f'{name}_hist.csv'),
                      index_col=0,
                      squeeze=True)
metrics_ss = pd.read_csv(os.path.join('metrics', f'{name}_metrics.csv'),
                         index_col=0,
                         squeeze=True)
example_df = pd.read_csv(os.path.join('metrics', f'{name}_examples.csv'))
```

```python
plt.figure(figsize=(20, 5))
hist_ss.plot.bar()
```

```python
plt.figure(figsize=(20, 5))
plt.xticks(np.around(np.arange(min(metrics_ss), max(metrics_ss) + 0.04, 0.04), 2))
metrics_ss.plot.line()
```

```python
bins = pd.qcut(example_df['purchases_30d__count'], q=10, duplicates='drop')
example_df.groupby(['w', bins]).agg({'y': 'mean',
                                     'uplift': 'mean'})
```

```python
bins = pd.qcut(example_df['purchases_sum_30d__sum'], q=10, duplicates='drop')
example_df.groupby(['w', bins]).agg({'y': 'mean',
                                     'uplift': 'mean'})
```

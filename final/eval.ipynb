```python
import os
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
name = 'loo'
```

```python
hist_ss = pd.read_csv(os.path.join('metrics', f'{name}_hist.csv'),
                      index_col=0,
                      squeeze=True)
metrics_df = pd.read_csv(os.path.join('metrics', f'{name}_metrics.csv'))
example_df = pd.read_csv(os.path.join('metrics', f'{name}_examples.csv'))
```

```python
plt.figure(figsize=(20, 5))
hist_ss.plot.bar()
```

```python
plt.figure(figsize=(20, 5))
plt.xticks(np.around(np.arange(min(metrics_df['k']), max(metrics_df['k']) + 0.02, 0.02), 2))
metrics_df.set_index('k')['mean'].plot.line()
```

```python
plt.figure(figsize=(20, 5))
metrics_df.set_index('k')['std'].plot.line()
```

```python
bins = pd.qcut(example_df['purchases_30d__count'], q=10, duplicates='drop')
example_df.groupby(['w', bins]).agg({'y': 'mean',
                                     'uplift': 'mean'})
```

```python
bins = pd.qcut(example_df['purchases_sum_30d__sum'], q=10, duplicates='drop')
example_df.groupby(['w', bins]).agg({'y': 'mean',
                                     'uplift': 'mean'})
```
